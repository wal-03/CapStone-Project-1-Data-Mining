```python
#ingin melihat keunikan data dari masing2 kolom
kolom_unik = ['Timestamp', 'Bidang wirausaha', 'Perusahaan', 'Posisi','Saran']
for kolom in list(set(df.columns) - set(kolom_unik)):
    print(f'Nama Kolom: {kolom}')
    try:
        for unik in sorted(df[kolom].unique(), key=lambda x: (isinstance(x, str), x)):
            print(unik)
    except:
        for unik in sorted([None if x is pd.NA else x for x in df[kolom].unique()], key=lambda x: float('inf') if x is None else x):
            print(unik)
    print('\n')
```

setelah ini akan dilakukan pengoreksian noise dan perubahan pada beberapa data di masing2 kolom yang tidak sesuai dengan data lainnya pada kolom yang sama

```python
#Kolom Tahun Wirausaha
#Nampak bahwa beberapa data yang bukan numerik
for unik in df['Tahun Wirausaha'].unique():
    if not f'{unik}'.isnumeric():
        df.loc[df['Tahun Wirausaha'] == unik, 'Tahun Wirausaha'] = np.nan

#Kolom Status saat ini
#Menyamakan kategori pada kolom ini
df['Status saat ini'] = df['Status saat ini'].replace(['Internship ','internship','Internship','Magang','mengikuti pelatihan'], 'Internship')
df['Status saat ini'] = df['Status saat ini'].replace(['Freelance', 'Free Lancer dan Co-Founder', 'Pekerja Harian Lepas', 'Part time','Content Creator'], 'Freelance')
df['Status saat ini'] = df['Status saat ini'].replace(['Belum mendapat pekerjaan', 'Menganggur', 'Akan mencari pekerjaan', 'Proses Wisuda'], 'sedang mencari kerja')
df['Status saat ini'] = df['Status saat ini'].replace(['Masih dalam pendidikan untuk persyaratan pekerjaan','Mendaftar studi magister','Mahasiswa', 'Melanjutkan Studi'], 'melanjutkan studi')
df['Status saat ini'] = df['Status saat ini'].replace(['Mendapatkan Pekerjaan', 'Karyawan ','Pegawai Tetap ', 'Staf', 'Operator Sekolah'], 'Karyawan Swasta')
df['Status saat ini'] = df['Status saat ini'].replace(['Pegawai tidak tetap instansi pemerintah'], 'Karyawan BUMN')
df['Status saat ini'] = df['Status saat ini'].replace(['Staff laboratorium ','Research Assistant'], 'Researcher')
df['Status saat ini'] = df['Status saat ini'].replace(['Pengajar les dan wirausaha', 'Mengelola Yayasan Pendidikan'], 'Pengajar')

#Kolom Tahun Masuk Studi Lanjut
#melakukan seleksi dna menisakan tahun masuk S2 saja
for unik in df['Tahun Masuk Studi Lanjut'].unique():
    if '-' in f'{unik}' or '—' in f'{unik}' or 'Baru' in f'{unik}':
        df.loc[df['Tahun Masuk Studi Lanjut'] == unik, 'Tahun Masuk Studi Lanjut'] = np.nan
    if '20' in f'{unik}' and ',' in f'{unik}':
        tahun = ''
        for elemen in unik:
            if len(tahun) == 4:
                break
            if elemen.isnumeric():
                tahun += elemen
            else:
                tahun = ''
        df.loc[df['Tahun Masuk Studi Lanjut'] == unik, 'Tahun Masuk Studi Lanjut'] = tahun

#Kolom Penghasilan
#Merubah semua data menjadi float dengan ketelitian satu angka di belakang koma
for unik in df['Penghasilan'].unique():
    if ',' in f'{unik}':
        if unik[0] == '3':
            df.loc[df['Penghasilan'] == unik, 'Penghasilan'] = '0.3'
        else:
            df.loc[df['Penghasilan'] == unik, 'Penghasilan'] = unik[:unik.index(',')+2].replace(',', '.')

#Kolom Tahun Wisuda
#Membenahi data yang salah melakukan Input
df.loc[df['Tahun Wisuda'] == '22 Oktober 2020', 'Tahun Wisuda'] = '2020'

#Kolom Tahun Masuk UIN Jakarta
#Membenahi data yang salah melakukan Input
df.loc[df['Tahun Masuk UIN Jakarta'] == '2019 Genap', 'Tahun Masuk UIN Jakarta'] = '2019'
df.loc[df['Tahun Masuk UIN Jakarta'] == '1.11809E+13', 'Tahun Masuk UIN Jakarta'] = '2020'
df.loc[df['Tahun Masuk UIN Jakarta'] == '206', 'Tahun Masuk UIN Jakarta'] = '2006'
df.loc[df['Tahun Masuk UIN Jakarta'] == '208', 'Tahun Masuk UIN Jakarta'] = '2018'

#Kolom Universitas
#Membenahi data yang salah melakukan Input
df['Universitas'] = df['Universitas'].replace([np.nan, '-', '5,000,000', '—', 'Baru mau mendaftar di semester genap', 'Belum', 'Kota Tangerang'], np.nan)

#Kolom Nama Prodi S2/S3
#Membenahi data yang salah melakukan Input
for unik in df['Nama Prodi S2/S3'].unique():
    if len(f'{unik}') <= 2 or 'Baru' in f'{unik}':
        df.loc[df['Nama Prodi S2/S3'] == unik, 'Nama Prodi S2/S3'] = np.nan

#Kolom Program Studi
#Menyamakan kategori pada kolom ini
df['Program Studi'] = df['Program Studi'].replace(['Kimia'], 'Prodi Kimia')
df['Program Studi'] = df['Program Studi'].replace(['Biologi'], 'Prodi Biologi')
df['Program Studi'] = df['Program Studi'].replace(['Matematika'], 'Prodi Matematika')
df['Program Studi'] = df['Program Studi'].replace(['Agribisnis S1', 'Prodi Agribisnis'], 'Prodi Agribisnis S1')
df['Program Studi'] = df['Program Studi'].replace(['Prodi TI', 'Teknik Informatika'], 'Prodi Teknik Informatika')
df['Program Studi'] = df['Program Studi'].replace(['Prodi MAG', 'Agribisnis S2'], 'Prodi Agribisnis S2')
df['Program Studi'] = df['Program Studi'].replace(['Fisika'], 'Prodi Fisika')

#Kolom Waktu Mendapatkan Pekerjaan
#Membenahi salah inputasi
for unik in [i for i in df['Waktu Mendapatkan Pekerjaan'].unique() if len(i)>2]:
    if '>' in unik or 'lebih dari' in unik:
        df.loc[df['Waktu Mendapatkan Pekerjaan'] == unik, 'Waktu Mendapatkan Pekerjaan'] = 13
    else:
        bulan = ''
        for elemen in unik:
            if not elemen.isnumeric() and len(bulan) != 0:
                break
            if elemen.isnumeric():
                bulan += elemen
        if bulan:
            df.loc[df['Waktu Mendapatkan Pekerjaan'] == unik, 'Waktu Mendapatkan Pekerjaan'] = bulan
        else:
            df.loc[df['Waktu Mendapatkan Pekerjaan'] == unik, 'Waktu Mendapatkan Pekerjaan'] = np.nan

for kolom in kolom_unik:
    df[kolom] = df[kolom].replace([np.nan, '-', '—'], np.nan)
```

```python
#melakukan perubahan tipe data
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Tahun Masuk UIN Jakarta'] = df['Tahun Masuk UIN Jakarta'].astype('Int64')
df['Tahun Wisuda'] = df['Tahun Wisuda'].astype('Int64')
df['Bulan Wisuda'] = df['Bulan Wisuda'].astype('category')
df['Waktu Mendapatkan Pekerjaan'] = df['Waktu Mendapatkan Pekerjaan'].astype('Float64')
df['Waktu Mendapatkan Pekerjaan'] = df['Waktu Mendapatkan Pekerjaan'].astype('Int64')
df['Status saat ini'] = df['Status saat ini'].astype('category')
df['Tahun Wirausaha'] = df['Tahun Wirausaha'].astype('Int64')
df['Penghasilan'] = df['Penghasilan'].astype('float')
df['Penghasilan'] = df['Penghasilan']*1000000
df['Penghasilan'] = df['Penghasilan'].astype('Int64')
df['Universitas'] = df['Universitas'].astype('category')
df['Tahun Masuk Studi Lanjut'] = df['Tahun Masuk Studi Lanjut'].astype('Int64')
df['Nama Prodi S2/S3'] = df['Nama Prodi S2/S3'].astype('category')
df['Kesesuaian Kurikulum'] = df['Kesesuaian Kurikulum'].astype('Int64')
df['Kesesuaian Bidang Ilmu'] = df['Kesesuaian Bidang Ilmu'].astype('Int64')
df['Program Studi'] = df['Program Studi'].astype('category')
```

```python
# Dictionary untuk menyamakan nama universitas
universitas_mapping = {
    'IPB ': 'IPB',
    'IPB': 'IPB',
    'IPB University': 'IPB',
    'Institut Pertanian Bogor': 'IPB',
    'ITB Bandung': 'ITB',
    'Institut Teknologi Bandung': 'ITB',
    'UGM': 'UGM',
    'Universitas Gadjah Mada': 'UGM',
    'Universitas Indonesia': 'UI',
    'Universitas Indonesia (Depok)': 'UI',
    'Universitas Indonesia, Depok': 'UI',
    'Uin': 'UIN',
    'Universitas Andalas, Padang ': 'Universitas Andalas',
    'Universitas Krisnadwipanaya (UNKRIS)': 'UNKRIS',
    'Kwik Kian Gie School of Business (Jakarta)': 'Kwik Kian Gie School of Business',
    'Politeknik Elektronika Negeri Surabaya': 'Politeknik Elektronika Negeri Surabaya',
    'Innopolis University, Tatarstan, Russia': 'Innopolis University',
    'Universitas Mercubuana': 'Universitas Mercubuana'}

# Mengganti nama universitas berdasarkan dictionary
df['Universitas'] = df['Universitas'].replace(universitas_mapping)

#membuat kolom baru yang berisi univeristas S3
df['Universitas S3'] = df['Universitas'].apply(lambda x: x.split(',', 1)[1].strip() if isinstance(x, str) and ',' in x else np.nan)
df['Universitas'] = df['Universitas'].apply(lambda x: x.split(',', 1)[0].strip() if isinstance(x, str) and ',' in x else x)

#kembali melakukan penyederhanaan nama
df['Universitas'] = df['Universitas'].replace({'ITB Bandung': "ITB", 'S2: Universitas Indonesia':'UI'})
df['Universitas S3'] = df['Universitas S3'].replace({'S3: Institut Teknologi Bandung': "ITB"})

#melakukan pengecekan ulang nama2 yang sudah disederhanakan
for unik in sorted(df['Universitas'].unique(), key=lambda x: (isinstance(x, str), x)):
    print(unik)
print(df["Universitas S3"].unique())
```

```python
#membuat kolom baru yang berisi tahun masuk S3 mahasiswa
def tahun_masuk_s3(s3):
    if s3['Universitas S3'] == 'University of Groningen':
        return 2024
    if s3['Universitas S3'] == 'UGM':
        return 2019
    if s3['Universitas S3'] == 'ITB':
        return 2023

df['Tahun Masuk S3'] = df.apply(tahun_masuk_s3, axis=1)
```
